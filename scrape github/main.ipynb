```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
url = "https://github.com/topics"
response = requests.get(url)
contents = response.text
soup = BeautifulSoup(contents,'html.parser')
topic_names = soup.find_all('p',class_='f3 lh-condensed mb-0 mt-1 Link--primary')
desc = soup.find_all('p',class_='f5 color-fg-muted mb-0 mt-1')
topic_link = soup.find_all('a',class_='no-underline flex-1 d-flex flex-column')

topic_name_list = []
for  i in topic_names:
    topic_name_list.append(i.text.strip())
    
desc_list = []
for  i in desc:
    desc_list.append(i.text.strip())
    
url_list = []
base_url = 'https://github.com'
for i in topic_link:
    url_list.append(base_url + i['href'])

dictt = {'name':topic_name_list,'description':desc_list,'url':url_list}
df = pd.DataFrame(dictt)
df.to_csv('topic_list.csv', index=False)
df
```

```python
url2 = "https://github.com/topics/3d"
response2 = requests.get(url2)
contents2 = response2.text
soup = BeautifulSoup(contents2,'html.parser')
n1 = soup.find_all('h3', class_='f3 color-fg-muted text-normal lh-condensed')
m1 = soup.find_all('span', class_='Counter js-social-count')

username = []
for i in n1:
    n2 = i.find('a').text.strip()
    username.append(n2)
    
repository_name = []
for i in n1:
    n3 = i.find_all('a')[1].text.strip()
    repository_name.append(n3)

star = []
for i in m1:
    star.append(i.text.strip())

dictt = {'username':username,'repository_name':repository_name,'star':star}
df = pd.DataFrame(dictt)
df.to_csv('topic.csv', index=False)
df
```
