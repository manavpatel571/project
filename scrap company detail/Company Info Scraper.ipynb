* For multiple page

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML,like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

'''"This loop iterates through all the companies on the page and extracts information such as the name,
rating, reviews, company type, headquarters, year founded, and number of employees, if available. However,
due to the unstructured nature of the web page, some information might end up in the wrong cell. If the
number of employees is not listed on the page, it is recorded as NaN."'''

def scrape_page(page):
    url = f'https://www.ambitionbox.com/list-of-companies?page={page}'
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    companies = soup.find_all('div', class_='company-content')
    name = []
    rating = []
    reviews = []
    ctype = []
    hq = []
    old = []
    employee = []
    for i in companies:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('p',class_='rating').text.strip())
        reviews.append(i.find('a',class_='review-count').text.strip())
        ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
        hq.append(i.find_all('p', class_='infoEntity')[1].text.strip() if len(i.find_all('p', class_='infoEntity')) > 1 else np.nan)
        old.append(i.find_all('p',class_='infoEntity')[2].text.strip() if len(i.find_all('p', class_='infoEntity')) > 2 else np.nan)
        employee.append(i.find_all('p', class_='infoEntity')[3].text.strip() if len(i.find_all('p', class_='infoEntity')) > 3 else np.nan)
    d = {'name':name,'rating':rating,'reviews':reviews,'ctype':ctype,'hq':hq,'old':old,'employee':employee}
    df = pd.DataFrame(d)
    return df


dfs = []
for page in range(1, 6):
    df = scrape_page(page)
    dfs.append(df)

result = pd.concat(dfs).reset_index(drop=True)
result.to_csv('company_data.csv', index='False')
result
```
